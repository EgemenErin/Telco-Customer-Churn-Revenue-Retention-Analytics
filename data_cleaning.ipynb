```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("telcochurn.csv")
```

```python
df.head()
```

```python
print(df.shape)
print(df.dtypes)
```

```python
print(df.info)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(0)
print(df.isnull().sum())
```

```python
import numpy as np
df = df.replace(r'^\s*$', np.nan, regex=True)
print(df.isnull().sum())
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
print(f"Overall Churn Rate: {df['Churn'].mean():.2%}")
```

```python
print(df["tenure"])
```

```python
bins = [0, 12, 24, 36, 48, 60, 72]
labels = ['0-12 Months', '13-24 Months', '25-36 Months', '37-48 Months', '49-60 Months', '61-72 Months']
df['Tenure_Cohort'] = pd.cut(df['tenure'], bins=bins, labels=labels, include_lowest=True)
print(df['Tenure_Cohort'].value_counts())
```
